```python
import pandas as pd

import pydotplus
adult=pd.read_csv("C:\\Users\\a8904\\OneDrive\\桌面\\資料探勘作業一\\資料探勘作業一\\adult.csv",sep='\t')

adult
```

```python
from sklearn.preprocessing  import LabelEncoder

le=LabelEncoder()

adult['workclass']=le.fit_transform(adult['workclass'])
adult['education']=le.fit_transform(adult['education'])
adult['marital-status']=le.fit_transform(adult['marital-status'])
adult['occupation']=le.fit_transform(adult['occupation'])
adult['relationship']=le.fit_transform(adult['relationship'])
adult['race']=le.fit_transform(adult['race'])
adult['sex']=le.fit_transform(adult['sex'])
adult['native-country']=le.fit_transform(adult['native-country'])
adult['output'] = le.fit_transform(adult['output'])
adult
```

```python
x=adult.drop(['output'],axis=1)
x
```

```python
y=adult[['output']]
y
```

```python
from sklearn import tree 
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True)

clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, y_train)

```

```python
predict_value = clf.predict(x_test)
predict_value

finaldata=pd.DataFrame(predict_value)
finaldata
```

```python
finaldata['真實']=y
# finaldata=finaldata.rename(columns={'0': '預測'}, inplace=True)
finaldata
```

```python
finaldata.to_csv('C:\\Users\\a8904\\OneDrive\\桌面\\資料探勘作業一\\資料探勘作業一\\分類結果.csv',encoding=('utf-8-sig'))
```

```python
clf.score(x_train,y_train)

```

```python
clf.score(x_test,y_test)
```

```python
tree.plot_tree(clf)
```
