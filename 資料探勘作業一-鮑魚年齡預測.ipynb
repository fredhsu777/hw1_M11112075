```python
import pandas as pd
import matplotlib.pyplot as plt

a=pd.read_csv("C:/Users/a8904/OneDrive/桌面/資料探勘作業一/資料探勘作業一/鮑魚年齡.csv",sep=',')
                 
a
```

```python
from sklearn.preprocessing  import LabelEncoder

le=LabelEncoder()


a['Sex']=le.fit_transform(a['Sex'])

a
```

```python
x=a.drop(['Rings'],axis=1)
x
```

```python
y=a[['Rings']]

y
```

```python
from sklearn import tree 
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True)

clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, y_train)
```

```python

predict_value = clf.predict(x_test)
predict_value

finaldata=pd.DataFrame(predict_value)
finaldata
```

```python
finaldata['真實']=y

finaldata

```

```python
finaldata.to_csv('C:\\Users\\a8904\\OneDrive\\桌面\\資料探勘作業一\\資料探勘作業一\\鮑魚年齡分類結果.csv',encoding=('utf-8-sig'))
```

```python
clf.score(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
tree.plot_tree(clf)
```
